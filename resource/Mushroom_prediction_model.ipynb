### start modelling



```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier

from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.2,random_state=42)
model1=LogisticRegression()
model2=GaussianNB()
model3=SVC()
model4=KNeighborsClassifier( )
model5=DecisionTreeClassifier()
model6=RandomForestClassifier()
model7=AdaBoostClassifier()
```

```python
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)
model4.fit(x_train,y_train)
model5.fit(x_train,y_train)
model6.fit(x_train,y_train)
model7.fit(x_train,y_train)
```

```python
print('logistic accuracy: ',accuracy_score(y_test,model1.predict(x_test)))
print('naive accuracy: ',accuracy_score(y_test,model2.predict(x_test)))
print('svm accuracy: ',accuracy_score(y_test,model3.predict(x_test)))
print('knn accuracy: ',accuracy_score(y_test,model4.predict(x_test)))
print('decision accuracy: ',accuracy_score(y_test,model5.predict(x_test)))
print('random accuracy: ',accuracy_score(y_test,model6.predict(x_test)))
print('adaboost accuracy: ',accuracy_score(y_test,model7.predict(x_test)))
```

```python
print('\n\nlogistic matrix: \n',confusion_matrix(y_test,model1.predict(x_test)))
print('\n\nnaive matrix: \n',confusion_matrix(y_test,model2.predict(x_test)))
print('\n\nsvm matrix: \n',confusion_matrix(y_test,model3.predict(x_test)))
print('\n\nknn matrix: \n',confusion_matrix(y_test,model4.predict(x_test)))
print('\n\ndecision matrix: \n',confusion_matrix(y_test,model5.predict(x_test)))
print('\n\nrandom matrix: \n',confusion_matrix(y_test,model6.predict(x_test)))
print('\n\nadaboost matrix: \n',confusion_matrix(y_test,model7.predict(x_test)))
```
